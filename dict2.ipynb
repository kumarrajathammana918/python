```python
# Question 49: Create a dictionary with nested sets:
y= {
    'groups': {{1, 2, 3}, {4, 5, 6}},
     'categories': {{'a', 'b'}, {'c', 'd'}}}

print(y)
```

```python
# Question 50: Find union of all nested sets
x={
    'groups': {{1, 2, 3}, {4, 5, 6}},
     'categories': {{'a', 'b'}, {'c', 'd'}}}
y=set()
for i in x['groups']:
    y|=i
for k in x['categories']:
    y|=k
print(y)
```
