```python
# Question 2: Check if "racecar" is a palindrome

x="racecar"
rev=""
i=len(x)-1
while i>=0:
   rev+=x[i]
   i-=1
print("the reversed num is  ",rev)
if(rev==x) :
    print("it is s palindrome")
else :
  print("its  not a palindrome")  
```
